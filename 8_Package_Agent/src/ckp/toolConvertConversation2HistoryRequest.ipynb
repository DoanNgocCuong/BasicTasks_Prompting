Dưới đây là mô tả logic của mã code mà tôi đã viết:

### Logic Mã Code

1. **Đọc Dữ Liệu Từ File Excel**:
   - Sử dụng hàm `read_excel_file` để đọc dữ liệu từ file Excel. Chỉ giữ lại hai cột: "Role" và "Content".

2. **Khởi Tạo Dữ Liệu Đầu Ra**:
   - Tạo một từ điển `output_data` với các danh sách rỗng cho các cột đầu ra: `conversation_history`, `previous_assistant_response`, `user_answer`, và `next_assistant_response`.
   - Khởi tạo một danh sách `conversation_history` để lưu trữ lịch sử hội thoại.

3. **Lặp Qua Dữ Liệu**:
   - Sử dụng vòng lặp `for` để duyệt qua từng hàng trong DataFrame:
     - **Nếu `Role` là `roleA`**:
       - Lưu nội dung của `roleA` vào biến `user_content` và thêm nó vào danh sách `user_answer`.
       - Cập nhật `previous_assistant_response` bằng cách lấy nội dung của dòng `roleB` trước đó (nếu có).
       - Cập nhật `next_assistant_response` bằng cách lấy nội dung của dòng `roleB` sau đó (nếu có).
       - Cập nhật `conversation_history` bằng cách thêm các dòng trước đó (bao gồm cả `roleB`) nhưng không bao gồm dòng `roleA` hiện tại.
       - Thêm dòng `roleA` vào `conversation_history` cho các lần trước đó.

     - **Nếu `Role` là `roleB`**:
       - Thêm dòng `roleB` vào `conversation_history`.

4. **Cập Nhật `conversation_history`**:
   - Khi gặp dòng `roleA`, mã sẽ cập nhật `conversation_history` để không bao gồm dòng `roleA` hiện tại, chỉ giữ lại các dòng trước đó và dòng `roleB` tiếp theo.

5. **Kiểm Tra Độ Dài Danh Sách**:
   - Trước khi tạo DataFrame, mã sẽ kiểm tra độ dài của từng danh sách trong `output_data` để đảm bảo rằng tất cả các danh sách đều có cùng độ dài.

6. **Tạo DataFrame và Xuất Ra File Excel**:
   - Cuối cùng, mã sẽ tạo một DataFrame từ `output_data` và ghi nó vào file Excel đầu ra.

### Kết Luận
Mã code này giúp tổ chức và lưu trữ thông tin hội thoại giữa hai vai trò `roleA` và `roleB`, đồng thời đảm bảo rằng lịch sử hội thoại được cập nhật chính xác mà không bao gồm dòng `roleA` tại thời điểm hiện tại. Nếu bạn cần thêm thông tin hoặc có câu hỏi nào khác, hãy cho tôi biết!


````python
import pandas as pd

# Đọc file Excel
def read_excel_file(file_path):
    df = pd.read_excel(file_path)
    df = df[['Role', 'Content']]
    return df

# Tạo file Excel mới với định dạng yêu cầu
def create_new_excel_file(df, output_file_path):
    output_data = {
        "conversation_history": [],
        "previous_assistant_response": [],
        "user_answer": [],
        "next_assistant_response": []
    }
    
    conversation_history = []  # Khởi tạo lịch sử hội thoại

    for index, row in df.iterrows():
        # print(f"Processing row {index}: {row['Role']} - {row['Content']}")  # In ra thông tin hàng hiện tại

        if row['Role'] == 'roleA':
            user_content = row["Content"]
            output_data["user_answer"].append(user_content)

            # Cập nhật previous_assistant_response
            if index > 0 and df.iloc[index - 1]['Role'] == 'roleB':
                previous_assistant_response = df.iloc[index - 1]["Content"]
            else:
                previous_assistant_response = ""
            output_data["previous_assistant_response"].append(previous_assistant_response)

            # Cập nhật next_assistant_response
            if index + 1 < len(df) and df.iloc[index + 1]['Role'] == 'roleB':
                next_assistant_response = df.iloc[index + 1]["Content"]
            else:
                next_assistant_response = ""
            output_data["next_assistant_response"].append(next_assistant_response)

            # Cập nhật conversation_history, bỏ dòng roleA hiện tại
            # output_data["conversation_history"].append(
            #     f'[\n    ' + ',\n    '.join(conversation_history) + '\n    ' + f'{{"role": "roleB", "content": "{df.iloc[index + 1]["Content"]}"}}' + '\n]'
            # ) 
            
            output_data["conversation_history"].append(
                '[\n    ' + ',\n    '.join(conversation_history) + ',\n    ' + f'{{"role": "roleB", "content": "{df.iloc[index + 1]["Content"]}"}}\n]'
            )           
            print(output_data)
            # Thêm dòng roleA vào conversation_history cho các lần trước
            conversation_history.append(f'{{"role": "{row["Role"]}", "content": "{user_content}"}}')

        elif row['Role'] == 'roleB':
            # Thêm roleB vào conversation_history
            conversation_history.append(f'{{"role": "{row["Role"]}", "content": "{row["Content"]}"}}')

        elif row['Role'] == 'Separator':
            # Append empty strings for the end of conversation
            output_data["conversation_history"].append('--- End of Conversation ---')
            output_data["previous_assistant_response"].append("")
            output_data["user_answer"].append("")
            output_data["next_assistant_response"].append("")
            conversation_history = []  # Reset conversation history
            continue  # Skip to the next iteration

        # Cập nhật conversation_history cho mỗi dòng roleA
        if row['Role'] == 'roleA':
            output_data["conversation_history"][-1] = f'[\n    ' + ',\n    '.join(conversation_history[:-1]) + '\n]'  # Bỏ dòng roleA hiện tại

    # Kiểm tra độ dài của các danh sách
    print("Lengths of output_data lists:")
    for key in output_data:
        print(f"{key}: {len(output_data[key])}")

    # Đảm bảo tất cả các danh sách đều có cùng độ dài
    max_length = max(len(output_data["conversation_history"]),
                     len(output_data["previous_assistant_response"]),
                     len(output_data["user_answer"]),
                     len(output_data["next_assistant_response"]))

    for key in output_data:
        while len(output_data[key]) < max_length:
            output_data[key].append("")  # Thêm chuỗi rỗng nếu cần

    output_df = pd.DataFrame(output_data)
    output_df.to_excel(output_file_path, index=False)

# Đường dẫn đến file Excel cũ và file mới
input_file_path = './id18.xlsx'
output_file_path = './id18_output.xlsx'

# Đọc dữ liệu từ file cũ
df = read_excel_file(input_file_path)

# Tạo file Excel mới
create_new_excel_file(df, output_file_path)
````

````python
import pandas as pd
import json

# Đọc file Excel
def read_excel_file(file_path):
    df = pd.read_excel(file_path)
    df = df[['Role', 'Content', 'Full Log']]
    return df

# Tạo file Excel mới với định dạng yêu cầu
def create_new_excel_file(df, output_file_path):
    output_data = {
        "conversation_history": [],
        "previous_assistant_response": [],
        "user_answer": [],
        "next_assistant_response": [],
        "full_log": [],
        "INTENT_PREDICT_LLM": [],
        "CUR_INTENT": []
    }
    
    conversation_history = []

    for index, row in df.iterrows():
        if row['Role'] == 'roleA':
            user_content = row["Content"]
            output_data["user_answer"].append(user_content)

            # Cập nhật previous_assistant_response
            if index > 0 and df.iloc[index - 1]['Role'] == 'roleB':
                previous_assistant_response = df.iloc[index - 1]["Content"]
            else:
                previous_assistant_response = ""
            output_data["previous_assistant_response"].append(previous_assistant_response)

            # Cập nhật next_assistant_response, full_log và các intent
            if index + 1 < len(df) and df.iloc[index + 1]['Role'] == 'roleB':
                next_assistant_response = df.iloc[index + 1]["Content"]
                full_log = df.iloc[index + 1]["Full Log"] if not pd.isna(df.iloc[index + 1]["Full Log"]) else ""
                
                # Lấy intent từ full_log
                try:
                    if full_log:
                        full_log_json = json.loads(full_log)
                        record = full_log_json.get('logs', {}).get('record', {})
                        intent_predict = record.get('INTENT_PREDICT_LLM', '')
                        cur_intent = record.get('CUR_INTENT', '')
                    else:
                        intent_predict = ''
                        cur_intent = ''
                except (json.JSONDecodeError, AttributeError, KeyError):
                    intent_predict = ''
                    cur_intent = ''
            else:
                next_assistant_response = ""
                full_log = ""
                intent_predict = ""
                cur_intent = ""
                
            output_data["next_assistant_response"].append(next_assistant_response)
            output_data["full_log"].append(full_log)
            output_data["INTENT_PREDICT_LLM"].append(intent_predict)
            output_data["CUR_INTENT"].append(cur_intent)

            # Cập nhật conversation_history
            output_data["conversation_history"].append(
                '[\n    ' + ',\n    '.join(conversation_history) + ',\n    ' + f'{{"role": "roleB", "content": "{df.iloc[index + 1]["Content"]}"}}\n]'
            )
            
            conversation_history.append(f'{{"role": "{row["Role"]}", "content": "{user_content}"}}')

        elif row['Role'] == 'roleB':
            conversation_history.append(f'{{"role": "{row["Role"]}", "content": "{row["Content"]}"}}')

        elif row['Role'] == 'Separator':
            output_data["conversation_history"].append('--- End of Conversation ---')
            output_data["previous_assistant_response"].append("")
            output_data["user_answer"].append("")
            output_data["next_assistant_response"].append("")
            output_data["full_log"].append("")
            output_data["INTENT_PREDICT_LLM"].append("")
            output_data["CUR_INTENT"].append("")
            conversation_history = []
            continue

        if row['Role'] == 'roleA':
            output_data["conversation_history"][-1] = f'[\n    ' + ',\n    '.join(conversation_history[:-1]) + '\n]'

    # Kiểm tra và cân bằng độ dài các danh sách
    max_length = max(len(value) for value in output_data.values())
    for key in output_data:
        while len(output_data[key]) < max_length:
            output_data[key].append("")

    output_df = pd.DataFrame(output_data)
    output_df.to_excel(output_file_path, index=False)

# Đường dẫn đến file Excel cũ và file mới
input_file_path = './id18.xlsx'
output_file_path = './id18_output.xlsx'

# Đọc dữ liệu từ file cũ
df = read_excel_file(input_file_path)

# Tạo file Excel mới
create_new_excel_file(df, output_file_path)
````
