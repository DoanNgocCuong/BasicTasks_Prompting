## SetUp LLMs và JSON output 

- Prompt: 
    - Lấy y nguyên luồng của Flow_Dang1_LISTENING.ipynb
    - JSON Output: `Ensure that the output is a valid JSON object without any additional formatting or keywords (such as JSON).`

````python
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def get_completion(prompt, model="gpt-4o", temperature=0):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message.content
````

````python
#claude_ai
import json

def generate_topics(job_position):
    """
    Generate 7 topics related to the given job position in both work and life contexts.
    
    Args:
    job_position (str): The job position to generate topics for.
    
    Returns:
    str: A JSON string containing the job position and 7 related topics.
    """
    prompt = f"""
    You are an expert in career development and work-life balance.
    
    Task: Generate 7 topics related to the job position "{job_position}" that cover both professional and personal aspects.
    
    Instructions:
    1. Consider various aspects of the job including technical skills, soft skills, and lifestyle impacts.
    2. Ensure topics are relevant to both work situations and personal life.
    3. Make topics specific enough to be meaningful, but general enough to apply broadly within the field.
    4. Present the information in Vietnamese.
    5. Each topic MUST include the phrase "in industry {job_position}" or equivalent to emphasize relevance to the specific industry.
    6. Use specialized language and common terminology in the {job_position} field to make topics more specific and prominent.
    
    Return a JSON object with the following structure:
    (Ensure that the output is a valid JSON object without any additional formatting or keywords (such as JSON)
    
    {{
      "job_position": "{job_position}",
      "work_topics": {{
        "topic1": "Topic 1 description, for {job_position}",
        "topic2": "Topic 2 description, for ",
        "topic3": "Topic 3 description",
        "topic4": "Topic 4 description",
        "topic5": "Topic 5 description",
        "topic6": "Topic 6 description",
        "topic7": "Topic 7 description"
      }}
    }}
    """

    # Assuming you have a function called get_completion to generate the response
    response = get_completion(prompt)
    
    # Parse the JSON response
    try:
        topics_dict = json.loads(response)
        return json.dumps(topics_dict, ensure_ascii=False, indent=2)
    except json.JSONDecodeError:
        print("Error: The generated response is not valid JSON. Please check the output.")
        return None

# Example usage:
# result = generate_topics("IT")
# print(result)
````

````python
# claude_ai

def generate_situations(job_position, topic):
    """
    Generate 5 situations around the given topic for the specified job position.
    
    Args:
    job_position (str): The job position.
    topic (str): The topic to generate situations for.
    
    Returns:
    str: A JSON string containing the job position, topic, and 5 related situations.
    """
    prompt = f"""
    Bạn là chuyên gia về phát triển nghề nghiệp và tình huống công việc trong ngành {job_position}.
    
    Nhiệm vụ: Tạo ra 5 tình huống liên quan đến chủ đề "{topic}" cho vị trí {job_position}.
    
    Hướng dẫn:
    1. Tạo ra 5 tình huống cụ thể và thực tế liên quan đến chủ đề trong ngành {job_position}.
    2. Mỗi tình huống phải bao gồm:
       - Tên tình huống (tiếng Việt)
       - Chi tiết tình huống (tiếng Việt)
       - Vai trò của người dùng (character_me) (tiếng Việt)
       - Vai trò của AI (character_ai) (tiếng Việt)
       - Lời chào của AI (bằng tiếng Anh, bao gồm thông tin cá nhân và chào hỏi cơ bản)
       - Danh sách 8 câu hỏi liên quan đến tình huống (bằng tiếng Anh)
    3. Đảm bảo các tình huống đa dạng và phản ánh các khía cạnh khác nhau của chủ đề.
    4. Sử dụng ngôn ngữ chuyên ngành và thuật ngữ phổ biến trong lĩnh vực {job_position}.

    Trả về một đối tượng JSON với cấu trúc sau:
    (Ensure that the output is a valid JSON object without any additional formatting or keywords (such as JSON)
    {{
      "job_position": "{job_position}",
      "topic": "{topic}",
      "situations": [
        {{
          "situation_name": "Tên tình huống 1",
          "situation_details": "Chi tiết tình huống 1",
          "character_me": "Vai trò của người dùng",
          "character_ai": "Vai trò của AI",
          "AI_greeting": "Lời chào của AI bằng tiếng Anh",
          "questions_list": [
            "Câu hỏi 1",
            "Câu hỏi 2",
            "Câu hỏi 3",
            "Câu hỏi 4",
            "Câu hỏi 5",
            "Câu hỏi 6",
            "Câu hỏi 7",
            "Câu hỏi 8"
          ]
        }},
        ... (4 tình huống khác với cấu trúc tương tự)
      ]
    }}
    """

    # Assuming you have a function called get_completion to generate the response
    response = get_completion(prompt)
    
    # Parse the JSON response
    try:
        situations_dict = json.loads(response)
        return json.dumps(situations_dict, ensure_ascii=False, indent=2)
    except json.JSONDecodeError:
        print("Error: The generated response is not valid JSON. Please check the output.")
        return None

# Example usage:
# Step 1: Generate topics
# result = generate_topics("IT")
# work_topics_js = json.loads(result)
# topic1 = work_topics_js["work_topics"]["topic1"]

# Step 2: Generate situations for a specific topic
# situations = generate_situations("IT", topic1)
# print(situations)
````

````python

import json
import pandas as pd
import random

def process_excel_file(input_file_path, output_file_path):
    print(f"Bắt đầu xử lý file Excel: {input_file_path}")
    
    # Đọc file Excel đầu vào
    print("Đang đọc file Excel đầu vào...")
    df = pd.read_excel(input_file_path, sheet_name="LIST CHỦ ĐỀ")
    print(f"Đã đọc xong file Excel. Số hàng: {len(df)}")
    
    # Tạo DataFrame mới để lưu kết quả
    print("Khởi tạo DataFrame kết quả...")
    result_df = pd.DataFrame(columns=[
        "CÔNG VIỆC", "TOPIC", "TÊN TÌNH HUỐNG", "CHI TIẾT TÌNH HUỐNG", 
        "NHÂN VẬT TÔI", "NHÂN VẬT AI", "Giới tính AI", "Câu mở đầu của AI", 
        "LIST CÂU HỎI do AI hỏi"
    ])
    
    # Xử lý từng hàng trong DataFrame đầu vào
    total_rows = len(df)
    for index, row in df.iterrows():
        print(f"Đang xử lý hàng {index + 1}/{total_rows}")
        job_position = row["Vị trí"]
        print(f"Vị trí công việc: {job_position}")
        
        # Generate topics
        print(f"Đang tạo chủ đề cho {job_position}...")
        topics_json = generate_topics(job_position)
        topics_dict = json.loads(topics_json)
        print(f"Đã tạo xong {len(topics_dict['work_topics'])} chủ đề")
        
        total_topics = len(topics_dict["work_topics"])
        for topic_index, (topic_key, topic_value) in enumerate(topics_dict["work_topics"].items(), 1):
            print(f"Đang xử lý chủ đề {topic_index}/{total_topics}: {topic_value}")
            
            # Generate situations for each topic
            print("Đang tạo tình huống...")
            situations_json = generate_situations(job_position, topic_value)
            situations_dict = json.loads(situations_json)
            print(f"Đã tạo xong {len(situations_dict['situations'])} tình huống")
            
            for situation_index, situation in enumerate(situations_dict["situations"], 1):
                print(f"Đang xử lý tình huống {situation_index}/{len(situations_dict['situations'])}")
                print(f"Tên tình huống: {situation['situation_name']}")
                print(f"Chi tiết tình huống: {situation['situation_details']}")
                
                # Tạo một hàng mới cho DataFrame kết quả
                new_row = pd.DataFrame({
                    "CÔNG VIỆC": [job_position],
                    "TOPIC": [topic_value],
                    "TÊN TÌNH HUỐNG": [situation["situation_name"]],
                    "CHI TIẾT TÌNH HUỐNG": [situation["situation_details"]],
                    "NHÂN VẬT TÔI": [situation["character_me"]],
                    "NHÂN VẬT AI": [situation["character_ai"]],
                    "Giới tính AI": [random.choice(["Nam", "Nữ"])],  # Random gender
                    "LIST CÂU HỎI do AI hỏi": ["\n".join(situation["questions_list"])],
                    "Câu mở đầu của AI": [situation["AI_greeting"]]
                    
                })
                
                result_df = pd.concat([result_df, new_row], ignore_index=True)
                print(f"Đã thêm tình huống vào DataFrame kết quả. Tổng số hàng hiện tại: {len(result_df)}")
    
    # Lưu kết quả vào file Excel mới
    print(f"Đang lưu kết quả vào file: {output_file_path}")
    result_df.to_excel(output_file_path, index=False)
    print(f"Đã lưu kết quả thành công. Tổng số hàng trong file kết quả: {len(result_df)}")


# Sử dụng hàm
input_file_path = r"D:\OneDrive - Hanoi University of Science and Technology\ITE10-DS&AI-HUST\Learn&Task\PRODUCT_THECOACH\Task4_Topic_Situation_ContentGen\test_gen_topic_situation.xlsx"
output_file_path = "GenTopic_output_file.xlsx"
print("Bắt đầu quá trình xử lý...")
process_excel_file(input_file_path, output_file_path)
print("Quá trình xử lý đã hoàn tất.")
````
