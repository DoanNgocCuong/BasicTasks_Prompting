   - **Câu đặc biệt**: Cảm thán, Mệnh lệnh, Nghi vấn tu từ  

2. **Độ dài câu**:  
   - **Ngắn**: 1-3 từ  
   - **Trung bình**: 4-10 từ  
   - **Dài**: 11-15 từ  
   - **Rất dài**: >15 từ  

3. **Nền tảng ngôn ngữ**:  
   - **Tiếng Anh**  
   - **Tiếng Việt**  
   - **Song ngữ Anh-Việt**  

4. **Domain (Lĩnh vực & Đối tượng)**:

| **Danh mục chính**           | **Loại lỗi cụ thể**                                      | **Ví dụ**                     |
|------------------------------|----------------------------------------------------------|-------------------------------|
| **1. Lỗi âm đầu**            | /θ/→/t/, /ð/→/d/, /r/→/l/, /v/→/b/, /f/→/p/, /ʃ/→/s/    | *this*→*dis*, *fine*→*pine*   |
| **2. Lỗi âm giữa**           | Bỏ âm giữa, sai âm schwa /ə/, lỗi âm ghép /bl/, /kl/     | *comfortable*→*comftable*     |
| **3. Lỗi âm cuối**           | Bỏ âm cuối, thay âm cuối, âm gió không rõ               | *cat*→*ca*, *map*→*mat*       |

| **5. Lỗi ngữ điệu**          | Sai ngữ điệu câu hỏi/trần thuật, giọng đều              | *Are you okay?* (nói phẳng giọng)|
| **6. Lỗi nối & lược âm**     | Không nối âm, không lược âm, sai nối âm                 | *Get it*→*gedit*              |
