S·ª≠ d·ª•ng `dspy.MIPROv2` ƒë·ªÉ t·ªëi ∆∞u h√≥a prompt d·ª±a tr√™n h√†m ƒë√°nh gi√° ƒë·ªô ch√≠nh x√°c `answer_exact_match`.

#### **3.3. M√£ Python ƒê·∫ßy ƒê·ªß**

D∆∞·ªõi ƒë√¢y l√† m√£ Python ƒë·∫ßy ƒë·ªß ƒë·ªÉ c·∫•u h√¨nh DSPy, t·∫°o module Q&A, t·ªëi ∆∞u h√≥a prompt d·ª±a tr√™n d·ªØ li·ªáu hu·∫•n luy·ªán, v√† so s√°nh k·∫øt qu·∫£ tr∆∞·ªõc v√† sau khi t·ªëi ∆∞u h√≥a.

```python
import dspy
import pandas as pd
import json

# B∆∞·ªõc 1: C·∫•u h√¨nh m√¥ h√¨nh ng√¥n ng·ªØ (LM)
# Thay 'YOUR_OPENAI_API_KEY' b·∫±ng API key th·ª±c t·∫ø c·ªßa b·∫°n
lm = dspy.LM('openai/gpt-4o-mini', api_key='YOUR_OPENAI_API_KEY')
dspy.configure(lm=lm)

# B∆∞·ªõc 2: Chu·∫©n b·ªã t·∫≠p d·ªØ li·ªáu hu·∫•n luy·ªán
# Ch√∫ng ta ƒë√£ t·∫°o DataFrame 'data' ·ªü b∆∞·ªõc tr∆∞·ªõc
# Chuy·ªÉn ƒë·ªïi d·ªØ li·ªáu th√†nh danh s√°ch c√°c dictionary
trainset = [
    {"user_input": row["user_input"], "expected_output": row["expected_output"]}
    for _, row in data.iterrows()
]

# B∆∞·ªõc 3: T·∫°o m·ªôt module d·ª± ƒëo√°n
# ƒê·ªãnh nghƒ©a module v·ªõi input l√† 'user_input' v√† output l√† 'expected_output'
# S·ª≠ d·ª•ng ki·ªÉu d·ªØ li·ªáu 'str' cho c·∫£ input v√† output
qa_module = dspy.Predict("user_input: str -> expected_output: str")

# B∆∞·ªõc 4: ƒê√°nh gi√° hi·ªáu su·∫•t tr∆∞·ªõc khi t·ªëi ∆∞u h√≥a
print("=== Tr∆∞·ªõc khi t·ªëi ∆∞u h√≥a ===")
for example in trainset:
    prediction = qa_module(user_input=example["user_input"])
    print(f"C√¢u h·ªèi: {example['user_input']}")
    print(f"ƒê√°p √°n d·ª± ƒëo√°n: {prediction.expected_output}")
    print(f"ƒê√°p √°n mong mu·ªën: {example['expected_output']}")
    print("---")

# B∆∞·ªõc 5: C·∫•u h√¨nh thu·∫≠t to√°n t·ªëi ∆∞u h√≥a (v√≠ d·ª•: MIPROv2)
optimizer = dspy.MIPROv2(
    metric=dspy.evaluate.answer_exact_match,  # H√†m ƒë√°nh gi√° ƒë·ªô ch√≠nh x√°c
    auto="light"  # Ch·∫ø ƒë·ªô t·ªëi ∆∞u nh·∫π nh√†ng ƒë·ªÉ ti·∫øt ki·ªám chi ph√≠
)

# B∆∞·ªõc 6: Ti·∫øn h√†nh t·ªëi ∆∞u h√≥a
optimized_module = optimizer.compile(
    qa_module,
    trainset=trainset,
)

# B∆∞·ªõc 7: ƒê√°nh gi√° hi·ªáu su·∫•t sau khi t·ªëi ∆∞u h√≥a
print("\n=== Sau khi t·ªëi ∆∞u h√≥a ===")
for example in trainset:
    prediction = optimized_module(user_input=example["user_input"])
    print(f"C√¢u h·ªèi: {example['user_input']}")
    print(f"ƒê√°p √°n d·ª± ƒëo√°n: {prediction.expected_output}")
    print(f"ƒê√°p √°n mong mu·ªën: {example['expected_output']}")
    print("---")

# B∆∞·ªõc 8: S·ª≠ d·ª•ng module t·ªëi ∆∞u h√≥a cho c√¢u h·ªèi m·ªõi
new_question = "Social media is essential for modern communication. How does it benefit users?"
optimized_prediction = optimized_module(user_input=new_question)
print("\n=== C√¢u h·ªèi m·ªõi sau t·ªëi ∆∞u ===")
print(f"C√¢u h·ªèi: {new_question}")
print(f"ƒê√°p √°n d·ª± ƒëo√°n: {optimized_prediction.expected_output}")
```
