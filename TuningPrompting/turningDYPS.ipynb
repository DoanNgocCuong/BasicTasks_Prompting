### **3. Viết Prompt Cho Tác Vụ**

Trong tác vụ này, chúng ta muốn mô hình nhận **`user_input`** (đáp án của người dùng) và tạo **`expected_output`** (đáp án mong muốn dưới dạng JSON với các cải tiến). Để làm điều này, chúng ta sẽ sử dụng **DSPy** để huấn luyện mô hình dựa trên các ví dụ đã cung cấp.

#### **3.1. Định nghĩa Module DSPy**

Chúng ta sẽ sử dụng `dspy.Predict` để định nghĩa một module dự đoán `expected_output` dựa trên `user_input`.

#### **3.2. Cấu hình Thuật Toán Tối Ưu hóa**

Sử dụng `dspy.MIPROv2` để tối ưu hóa prompt dựa trên hàm đánh giá độ chính xác `answer_exact_match`.

#### **3.3. Mã Python Đầy Đủ**

Dưới đây là mã Python đầy đủ để cấu hình DSPy, tạo module Q&A, tối ưu hóa prompt dựa trên dữ liệu huấn luyện, và so sánh kết quả trước và sau khi tối ưu hóa.

```python
import dspy
import pandas as pd
import json

# Bước 1: Cấu hình mô hình ngôn ngữ (LM)
# Thay 'YOUR_OPENAI_API_KEY' bằng API key thực tế của bạn
lm = dspy.LM('openai/gpt-4o-mini', api_key='YOUR_OPENAI_API_KEY')
dspy.configure(lm=lm)

# Bước 2: Chuẩn bị tập dữ liệu huấn luyện
# Chúng ta đã tạo DataFrame 'data' ở bước trước
# Chuyển đổi dữ liệu thành danh sách các dictionary
trainset = [
    {"user_input": row["user_input"], "expected_output": row["expected_output"]}
    for _, row in data.iterrows()
]

# Bước 3: Tạo một module dự đoán
# Định nghĩa module với input là 'user_input' và output là 'expected_output'
# Sử dụng kiểu dữ liệu 'str' cho cả input và output
qa_module = dspy.Predict("user_input: str -> expected_output: str")

# Bước 4: Đánh giá hiệu suất trước khi tối ưu hóa
print("=== Trước khi tối ưu hóa ===")
for example in trainset:
    prediction = qa_module(user_input=example["user_input"])
    print(f"Câu hỏi: {example['user_input']}")
    print(f"Đáp án dự đoán: {prediction.expected_output}")
    print(f"Đáp án mong muốn: {example['expected_output']}")
    print("---")

# Bước 5: Cấu hình thuật toán tối ưu hóa (ví dụ: MIPROv2)
optimizer = dspy.MIPROv2(
    metric=dspy.evaluate.answer_exact_match,  # Hàm đánh giá độ chính xác
    auto="light"  # Chế độ tối ưu nhẹ nhàng để tiết kiệm chi phí
)

# Bước 6: Tiến hành tối ưu hóa
optimized_module = optimizer.compile(
    qa_module,
    trainset=trainset,
)

# Bước 7: Đánh giá hiệu suất sau khi tối ưu hóa
print("\n=== Sau khi tối ưu hóa ===")
for example in trainset:
    prediction = optimized_module(user_input=example["user_input"])
    print(f"Câu hỏi: {example['user_input']}")
    print(f"Đáp án dự đoán: {prediction.expected_output}")
    print(f"Đáp án mong muốn: {example['expected_output']}")
    print("---")

# Bước 8: Sử dụng module tối ưu hóa cho câu hỏi mới
new_question = "Social media is essential for modern communication. How does it benefit users?"
optimized_prediction = optimized_module(user_input=new_question)
print("\n=== Câu hỏi mới sau tối ưu ===")
print(f"Câu hỏi: {new_question}")
print(f"Đáp án dự đoán: {optimized_prediction.expected_output}")
```

### **5. Kết quả Kỳ vọng**

Khi chạy đoạn mã trên, bạn sẽ thấy sự cải thiện trong việc dự đoán `expected_output` dựa trên `user_input`. Cụ thể:

- **Trước khi tối ưu hóa:** Đáp án dự đoán có thể chưa chính xác hoặc không đầy đủ.
